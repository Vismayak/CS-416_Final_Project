## Data Cleaning and Extraction

```python
import pandas as pd
```

```python
raw_data = pd.read_csv("owid-energy-data.csv")
print(raw_data.shape)
```

```python
raw_data.head()
```

```python
# Keep only data of countries
country_data = raw_data[raw_data['iso_code'].notna()]
```

```python
#Data of countries 
world_data = raw_data[raw_data['country'] == "World"]
# Remove data before renewable energy data is available
world_data = world_data[world_data["renewables_share_elec"].notna()]
country_data = country_data[country_data["renewables_share_elec"].notna()]
# convert column to string
world_data["year"] = world_data["year"].astype(str)
print(len(world_data))
```

```python
codes = pd.read_csv("owid-energy-codebook.csv")
codes = codes[["column","description"]]
```

```python
world_data.to_csv("world_energy_data.csv", index=False)
country_data.to_csv("country_energy_data.csv", index=False)
```
